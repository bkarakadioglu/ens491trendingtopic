```python
import os, sys
import glob
import gzip
import json
from datetime import datetime

from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
from matplotlib.colors import LogNorm
```

```python
FIGLANG = "tr"
figtext = {
    "ylabel": {"tr": "Günlük değişim", "en": "Daily change"},
    "title": {"tr": "Örnek", "en": "Example"},
}

fig, axArr = plt.subplots(1, 4, figsize=(12,3), tight_layout=True)
axArr[0].set_ylabel(figtext["ylabel"][FIGLANG], fontsize=16)

sampleData = [
    [85, 25, 100, 50, 125, 950, 250, 95, 120, 55],
    [200, 220, 190, 210, 205, 195, 225, 190, 200],
    [85, 25, 100, 50, 750, -300, -250, -150, -100, 10, -5],
    [20, 400, -350, -250, -100, 350, 70, 700, -100, -50]
]

for i in range(len(sampleData)):
    markerline, stemlines, baseline = axArr[i].stem(sampleData[i])
    plt.setp(stemlines, 'color', '0.5')
    plt.setp(markerline, 'color', '0.5')
    axArr[i].set_xticks([])
    axArr[i].set_title('{} #{}'.format(figtext["title"][FIGLANG], i+1), fontsize=16)
plt.savefig('sample-charts-{}.png'.format(FIGLANG), dpi=300, transparent=True, bbox_inches='tight')    
```
